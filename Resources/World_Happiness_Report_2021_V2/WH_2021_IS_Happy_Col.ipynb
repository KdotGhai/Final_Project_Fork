```python
# read data
wh_report = "2021.csv"

df = pd.read_csv(wh_report)
```

```python
# Display data(orignal)
df.head()
```

```python
# Fix names of certain countries
df["Country name"].replace("Palestinian Territories","Palestine",inplace = True)
df["Country name"].replace("Hong Kong S.A.R. of China","Hong Kong",inplace = True)
df["Country name"].replace("Bosnia and Herzegovina"," Bosnia",inplace = True)
df["Country name"].replace("Taiwan Province of China","Taiwan",inplace = True)
```

```python
# do a df.describe() on "Ladder score"
df["Ladder score"].describe()
```

```python
df["Ladder score"].value_counts()
```

```python
# create "IS_HAPPY" column (binary outputs [Yes:1, No:0])

happy = []

for value in df["Ladder score"]:
    if value >= 5.5:
        happy.append(1)
    elif value < 5.5:
        happy.append(0)

# Add column to dataframe then inspect afterwards

df["IS_HAPPY"] = happy

df.head()
```

```python
df["IS_HAPPY"].value_counts()
```

```python
# The scope of these changes made to
# pandas settings are local to with statement.
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,  # The "None is a place holder for integer values if wanting to set constraints on display"
                       'display.precision', 3,
                       ):
    print(df[["Country name","Ladder score", "IS_HAPPY"]])
```
